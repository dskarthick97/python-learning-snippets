# I. Line Plot

```python
import matplotlib.pyplot as plt
```

- Specify the data to plot

    |x| | 1 | 2 | 3 |

    |y| | 2 | 4 | 1 |

```python
x = [1, 2, 3]
y = [2, 4, 1]
```

```python
# specify X and Y axis labels -- optional
plt.xlabel("x-axis")
plt.ylabel("y-axis")

# give title to your graph -- optional
plt.title("my first graph")

# plot points on the graph
plt.plot(x, y)

# show the plot
plt.show()  # blocking function
```

# II. Customizing line plots

```python
x = [2, 2, 3, 4, 5, 6]
y = [2, 4, 3, 5, 2, 6]
```

```python
plt.style.available
```

```python
plt.style.use("fivethirtyeight")
```

```python
import numpy as np
```

```python
plt.plot(x, 
         y, 
         color="red", 
         linewidth=2, 
         linestyle="-.",
         marker="o",
         markersize=10,
         markerfacecolor="white"
)

# setting x and y axes limit
plt.xlim(0, 7)
plt.ylim(0, 7)

# setting the step value
plt.xticks(np.arange(0, 7, 0.5))

plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.title("Customize graph")

plt.show()
```

# III. Bar chart

```python
marks = {
    "English": 9,
    "Maths": 8,
    "Physics": 6,
    "Chemistry": 10,
    "Biology": 7
}
```

```python
marks
```

```python
ticks = range(1, 6)
```

```python
height = list(marks.values())
```

```python
height
```

```python
tick_label = list(marks.keys())
```

```python
plt.bar(ticks, height, tick_label=tick_label)
plt.show()
```

```python
# providing color and width to the bar

plt.bar(ticks, height, tick_label=tick_label, color=["r", "b"], width=0.5)
plt.show()
```

### Horizontal bar chart

```python
# here the width and height are opposite to vertical bar chart

plt.barh(ticks, height, tick_label=tick_label, color=["g"], height=0.7)
plt.show()
```

### Comparison bar chart

```python
mark_ = {
    "English": 10,
    "Maths": 7,
    "Physics": 8,
    "Chemistry": 10,
    "Biology": 9
}
```

```python
index = np.arange(5)
bar_width = 0.35
```

```python
plt.bar(index, height, label="Tony", width=bar_width)
plt.bar(index + bar_width, list(mark_.values()), label="Bruce", width=bar_width)

# placing the ticks in the middle
plt.xticks(index + bar_width / 2, tick_label)

plt.legend()
plt.show()
```

# IV. Histogram

> Diagram consisting of rectangles whose area is proportional to the frequency of a variable and whose width is equal to the class interval.

```python
ages = [2, 5, 70, 40, 30, 45, 50, 45, 43, 40, 44, 60, 7, 13, 57, 18, 90, 77, 32, 21, 20, 40]
```

```python
# set the variable range and no. of intervals
age_range = (0, 100)
bins = 10  # interval
```

```python
plt.hist(ages, bins, age_range, rwidth=0.90)
plt.xticks(range(0, 100, 10))
plt.show()
```

# V. Scatter Plot

---

A scatter plot is a two-dimensional data visualization that uses dots to represent the values obtained for two different variables - one plotted along the x-axis and the other plotted along the y-axis.

```python
x = list(range(1, 11))
y = [2, 4, 5, 7, 6, 8, 9, 11, 12, 12]
```

```python
plt.scatter(x, y)
plt.show()
```

```python
# customizing the scatter plot

plt.scatter(x, y, s=100, marker="*")
plt.show()
```
